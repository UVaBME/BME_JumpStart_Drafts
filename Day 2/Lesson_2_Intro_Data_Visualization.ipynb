```python
import numpy as np
import matplotlib.pyplot as plt
```

For this section, we will be "generating" data using `numpy` functions, such as `linspace` or `random.normal()`

```python
x = np.linspace(-5,5,num=100)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
np.random.seed(314159)
a = np.random.normal(0,2, size = 100)
b = np.random.normal(0,2, size = 100)

plt.scatter(a,b)
plt.show()
```

```python
np.random.seed(314159) #for reproducibility
data = [np.random.normal(0, std, size = 100) for std in range(1,4)]

bplot1 = plt.boxplot(data)
plt.show()
```

It is important to think of the display you want for data BEFORE you try plotting it. For example, plotting the data from our scatter plot and boxplots using the regular `plot` function gives weird results:

```python
plt.plot(a,b)
plt.show()
```

```python
plt.plot(data);
plt.show()
```

## Section 2: Modifying plotting options
It's useful when creating figures to customize features for easier interpretability. In this section, we will be focusing on useful `matplotlib` customizations.


### 2.1 Axis Labels

```python
x = np.linspace(-5,5,num=100)
y = x**2

plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Growth Rate')
plt.title('Growth Rate vs Time')

plt.show()
```

### 2.2 Data Marker Display
Most plotting functions will have a default display setting that can be changed. The parabola we have been displaying is actually comprised on 100 data points, that are being connected. Here, we can change the color and the shape of the marker.

```python
plt.plot(x, y, '.r') #'r' is for red, '.' is for shape
plt.show()
```

### 2.3 Overlaying Plots
As default, `matplotlib` will add plots to the same figure. For MATLAB users, this is the same as using `hold on` in MATLAB. When plotting multiple graphs, the colors will be changed by default.

```python
x = np.linspace(-5,5,num=100)
y = x**2
z = x

plt.plot(x,y)
plt.plot(x,z, linewidth = 5)

plt.show()
```

You can also overlay plots of different types onto the same figure. For example, our parabola and boxplots from earlier!

```python
plt.plot(x,y)
plt.boxplot(data)
plt.show()
```

### 2.4 Subplotting
Although it can be useful to visualize data on the same plot, we sometimes want to display different data or use different displays. The `subplot` function from `matplotlib` allows us to set the position of a plot within a figure.

```python
t = np.linspace(0,10, num = 100)
y1 = np.cos(t)
y2 = np.sin(t)

plt.subplot(2,1,1) #subplot(number_of_rows = 2, number_of_cols = 1, position = 1)
plt.plot(t,y1)
plt.subplot(2,1,2)
plt.plot(t,y2)

plt.show()
```

## 2.3 Seaborn Intro
`matplotlib` has great basic functionality and can be used very generally. Here, we are going to introduce `seaborn`, a package that makes some of the more complex plots easier to generate "prettier" graphs more intuitively. For those who have used R, it is very similar to ggplot2.

```python
import seaborn as sns

sns.boxplot(data = data)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.hist(data)

plt.subplot(1,2,2)
sns.histplot(data)
plt.show()
```
