#### Calculating aortic valve area (for loops)

Aortic valve area can be calculated as a function of systolic pressure (P) and cardiac output (CO, in L/min) as follows:
$ Av = \frac{CO}{\sqrt{P}} $

Assuming that the cardiac output is relatively constant, calculate the aortic valve area for systolic pressures ranging from 2 to 60mmHg.

*Hint: use the the power syntax to calculate the square root of a number: `x**0.5`*

```python
import math

#initialize variables
cardiac_output = 4
pressure = list(range(5, 61, 5))

#intialize empty list
AV = []

#write their code here
for p in pressure:
    area = cardiac_output/math.sqrt(p)
    AV.append(area)
```

#### Calculating GC Content (For loops and if statements)

You have performed DNA sequencing and obtained a sequence of nucleotides. You want to calculate the GC content of this sequence. GC content is the percentage of nucleotides that are either guanine or cytosine. How could you go about doing this?

```python
seq = 'TTGTTAGTGATGGCTAAATATAG'

#enter your solution here
gc_count = 0
for bp in seq:
    if bp == 'G' or bp == 'C':
        gc_count = gc_count + 1

gc_content = gc_count / len(seq)
gc_content
```

#### Identifying the Complementary DNA Strand (for loops and if statements)

Given a DNA sequence for a single DNA strand, we would like to identify the complementary DNA strand. Using a for loop identify our complementary strand to the provided sequence:

```python
seq = 'TTGTTA'

#Enter your solution here
comp_seq = ''

for bp in seq:
    if bp == 'A':
        comp_seq = comp_seq + 'T'
    elif bp == 'T':
        comp_seq = comp_seq + 'A'
    elif bp == 'G':
        comp_seq = comp_seq + 'C'
    elif bp == 'C':
        comp_seq = comp_seq + 'G'

comp_seq
```

#### Michaelis-Menten Kinetics (for loops, lists, plotting illustration)

Kinases are an enzyme that add a phosphate to an amino acid on another protein, called the substrate. Kinases follow Michaelis Menten kinetics, where the rate of the reaction is proportional to the concentration of the substrate. Assuming the concentration of enzyme is The rate of the reaction is given by the following equation:

$ v = \frac{V_{max}[S]}{K_a + [S]} $

```python
v_max = 80
Km = 20
concentration = [0, 10, 20, 30, 40, 50, 60, 70, 80, 100, 150, 200]

#Enter code solution here
v_list = []
for s in concentration:
    v = (v_max*s) / (Km + s)
    v_list.append(v)
v_list
```

```python
import matplotlib.pyplot as plt

#plot the Michaelis-Menten curve
plt.plot(concentration, v_list)

#add labels and titles
plt.xlabel('Concentration of Substrate ($\mu$M)')
plt.ylabel('Velocity ($\mu$M/min)')
plt.xlim([0,200])
plt.ylim([0,85])

#add point to Km
plt.scatter([Km], [v_max/2], color = 'k', s = 50, zorder = 10)
plt.plot([0, Km], [v_max/2, v_max/2], color = 'k', linestyle = '--', lw = 0.5)
plt.plot([Km, Km], [0, v_max/2], color = 'k', linestyle = '--', lw = 0.5)
plt.annotate('$K_M$', (Km + 1, v_max/4))
plt.annotate('$V_{max}/2$', (0, v_max/2 + 1))

#add Vmax line
plt.axhline(y = v_max, color = 'k', linestyle = '--', lw = 0.5)
plt.annotate('$V_{max}$', (185, v_max + 1))
```

#### Compute the distance between two points (for loops, lists, list comprehensions,numpy)

*from [Cam Mura's textbook here](https://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1004867)*
Exercise 8: Write a program to compute the distance, d(r1, r2), between two arbitrary (user-specified) points, r1 = (x1, y1, z1) and r2 = (x2, y2, z2), in 3D space. Use the usual Euclidean distance between two points—the straight-line, “as the bird flies” distance. 

$ d(r_1,r_2) = \sqrt{(x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2} $

 With your code in hand, note the ease with which you can adjust your entire data-analysis workflow simply by modifying a few lines of code that correspond to the definition of the distance function. 

```python
r1 = [3, 4, 5]
r2 = [4, 7, 8]

#option 1 with lists only
d = ((r1[0]-r2[0])**2 + (r1[1]-r2[1])**2 + (r1[2]-r2[2])**2)**0.5
print(d)

#option 2 with for loop
vector_sum = 0
for i,j  in zip(r1, r2):
    vector_sum = vector_sum + (i - j)**2

d = vector_sum**0.5
print(d)


#option 3 with list comprehension
d = sum([(i - j)**2 for i,j in zip(r1, r2)])**0.5
print(d)

#option 4 with numpy arrays
import numpy as np
r1 = np.array(r1)
r2 = np.array(r2)
d = np.sqrt(np.sum((r1 - r2)**2))
print(d)
```

#### Find the start codon in a sequence (while loops)

We would like to identify the location of the start codon in a transcript sequence. Use a while loop to identify the location of the start codon in the following sequence. The start codon is AUG.

```python
seq = 'TTGATCATGTTCGGCA'

i =0
#enter while loop here
while (seq[i:i+3] != 'ATG'):
    i = i + 1

#output the results!
print(f'The start codon ({seq[i:i+3]}) is found at the {i+1}th base pair')
```
